# Pipeline analisi mutazioni — utilizzo di utils.py
Basta cambiare `PATH` e i parametri, poi eseguire le celle in ordine.

```python
from utils import (
    generate_matrices,
    calculate_statistics,
    plot_volcanos,
    build_all_networks,
    calculate_metrics,
    compare_networks,
)

# ── Parametri principali ──────────────────────────────────────────────────
PATH        = "./kras_pancreas"   # cartella dati (input E output)
TARGET_GENE = "KRAS"

COOCC_PARAMS = {"p_val": 0.05, "log2or": 1.0, "min_cooc": 3}
ME_PARAMS    = {"p_val": 0.01, "log2or": -1.0}
# ─────────────────────────────────────────────────────────────────────────
```

```python
# 1. Matrici binarie e di co-occorrenza
generate_matrices(PATH, TARGET_GENE)
```

```python
# 2. Statistiche Fisher + FDR
calculate_statistics(PATH, TARGET_GENE, COOCC_PARAMS, ME_PARAMS)
```

```python
# 3. Volcano plots
plot_volcanos(PATH, TARGET_GENE, COOCC_PARAMS, ME_PARAMS)
```

```python
# 4. Reti statiche 2D (FULL + FILTERED 1-hop)
build_all_networks(PATH, TARGET_GENE, COOCC_PARAMS)
```

```python
# 5. Metriche topologiche
calculate_metrics(PATH, TARGET_GENE, COOCC_PARAMS)
```

```python
# 6. (Opzionale) Confronto fra due coorti
compare_networks(
    path_global   = "./pancreas",
    path_filtered = "./kras_pancreas",
    coocc_params_global   = COOCC_PARAMS,
    coocc_params_filtered = COOCC_PARAMS,
)
```
